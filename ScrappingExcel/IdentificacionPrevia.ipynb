```python
import pandas as pd
import os


Directory = r"C:\Users\fcolin\Desktop\Todos"
os.chdir(Directory)
files = os.listdir()
files = [f for f in files if f[-4:] == 'xlsx']
```

```python
# get the name of the sheets in an excel file
Sheets = []
for f in files:
    if f[-4:] == 'xlsx':
        sheets = pd.ExcelFile(f).sheet_names
        #print(sheets)
        Sheets.append(sheets)
        # 5 min 2019

#pd.read_excel(files[0], sheet_name=None).keys()
# separte a list of lists into a single list
Sheets2 = [item for sublist in Sheets for item in sublist]

df_sheets = pd.DataFrame(Sheets2)
df_sheets.drop_duplicates(inplace=True)
df_sheets
```

```python
# Selccionar manualmente
sheets_aceptadas  = ["ROPA DAMA", "SHASA MAN", "CALZADO", "BISUTERIA", "NO BISUTERIA", "BEAUTY"]
```

```python
#j = 0
hist_colnames = []

for j in range(len(files)):
    
    file = files[j]
    pestanas = pd.ExcelFile(file).sheet_names

    # coincidence beteween pestanas y sheets_aceptadas
    sheets_aceptadas_excel = []
    for pestana in pestanas: 
        if pestana in sheets_aceptadas:
            sheets_aceptadas_excel.append(pestana)
            
    for sheet in sheets_aceptadas_excel:
        file_excel = pd.read_excel(file, sheet_name=sheet)
        # get the index of the row with the max null values
        for i in range(len(file_excel.iloc[:,0].values)):
            if file_excel.iloc[i,0]=="No.":
                index = i
                break
        col_names = file_excel.iloc[index].values
        # rename the columns
        # file_excel.columns = col_names
        hist_colnames.append(col_names)

        # time: 9min
```

```python
# separte a list of lists into a single list
hist_colnames2 = [item for sublist in hist_colnames for item in sublist]
hist_colnames2 = pd.DataFrame(hist_colnames2).drop_duplicates()
pd.DataFrame(hist_colnames2).to_csv('hist_colnames2.csv',index=False)

# Asignado que columnas son las que se van a usar manualmente
```
